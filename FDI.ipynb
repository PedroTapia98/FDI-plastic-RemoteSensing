Cargando el dataset

```python
import rasterio
image_file = '/content/drive/MyDrive/Imagegg.tif'
satdat = rasterio.open(image_file)
```

```python
print(satdat)
print(satdat.name)
print(satdat.count)
```

```python
print(satdat.count)
print(satdat.indexes)
```

Bandas a utilizar

```python
b11 = satdat.read(11) #swir
b6 = satdat.read(6) #RE2
b4 = satdat.read(4) #red
b8 = satdat.read(8) #nir
b3 = satdat.read(3) #green
```

```python
from rasterio.plot import show
show(satdat )
```

Cálculo de índices


```python
#indices
ndvi = ((b8-b4)/(b8+b4))
ndwi = ((b3-b4)/(b3+b8))
nir1 = (b6 -(b11-b6)*((835.1-664.5)/(1613.7-664.5)))
fdi = (((b6 -(b11-b6)*((835.1-664.5)/(1613.7-664.5)))) - nir1)
```

```python
show(ndvi,cmap='RdYlGn')
```

```python
show(ndwi,cmap='RdYlGn')
```

```python
show(nir1,cmap='RdYlGn')
```

```python
show(fdi,cmap='RdYlGn')
```
